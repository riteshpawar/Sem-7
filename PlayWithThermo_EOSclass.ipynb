```python
def V(T,P):
    a=dart.Z(T,P)
    a=sc.array(a)   
    b=a*8314*T/P
    return b[0]

V(50,101325)
```

```python
def gRbyRT(T,P,Compound,eos):
    comp=perrysdata.Compound(Compound)
    dart=eosClass.EOS(comp)
    dart.setEOS(eos)
    Z=dart.Z(T,P)[0]
    gR=dart.gR(T,P,Z)
    gRbyRT=gR/(dart.R*T)
    return gRbyRT
def HR(T,P,Compound,eos,x):
    t1=T+x
    t2=T-x
    dgRbyRT=(gRbyRT(t1,P,Compound,eos)-gRbyRT(t2,P,Compound,eos))/(2*x)
    HR=-dgRbyRT*dart.R*T**2
    return HR
    

    
    
    
```
